# None 활용

```python
def add_and_maybe_multiply(a, b, c=None): # 더하기인데 어쩌면 곱하기까지 해준다
  result = a + b

  if c is not None:
    result = result * c

  return result
```

```python
add_and_maybe_multiply(5,3)
```

```python
add_and_maybe_multiply(5,3,10)
```

# for loops

```python
# None는 더하기 안 됨
sequence = [1, 2, None, 4, None, 5]
total = 0

for value in sequence:
  total += value
```

```python
# continue 사용해서 None 계산 안 하기
sequence = [1, 2, None, 4, None, 5]
total = 0

for value in sequence:
  if value is None:
    continue
  total += value
```

```python
total
```

```python
# break 사용해서 중단 시키기
sequence = [1, 2, 0, 4, 6, 5, 2, 1]
total_until_5 = 0
for value in sequence:
  if value == 5:
    break
  total_until_5 += value
```

```python
total_until_5
```

```python
for i in range(4):
  for j in range(4):
    if j > i:
      break
    print((i, j))
```

```python
for a, b, c in [[1,2,3],[4,5,6],[7,8,9]]:
  print(a,b,c)
```

# while

```python
x = 256
total = 0
while x > 0:
  if total > 500:
    break
  total += x
  x = x // 2
```
